```python
def prime(num):
  prime = 0
  for i in range(3,num):
    if i == 2:
     continue
    elif not num % i == 0:
      continue
    else:
      prime += 1
  if prime == 0:
    print(f"{num} is prime")
  else:
    print(f"{num} is not prime")
      
  
```

```python
def fact(num):
  factorial = 1
  if num == 0:
    print("Factorial is 1")
  elif num < 0:
    print(f"Factorial is not allowed for {num}")
  else:
   for i in range(1,num + 1):
     factorial = factorial * i
   return factorial 
     
```
